### series

```python
import pandas as pd
```

```python
import numpy as np
```

```python
s1=pd.Series([6,2,8,4],index=['a','b','c','d'])
```

```python
s1
```

```python
type(s1)
```

```python
s2=pd.Series(np.random.randint(1,7,10)) #1~6까지 10개의 난수 뽑기
```

```python
s2
```

```python
s2.value_counts() # 각 면의 개수를 알려줌 
```

```python
s2.value_counts().sort_index() #메소드인가 속성인가? / 동작인가 상태인가? => 액션, 메소드 즉 괄호가 필요 
```

```python
s2.value_counts().sort_values()
```

```python
s2.value_counts().sort_values(ascending=False) #역순으로 뽑기
```
