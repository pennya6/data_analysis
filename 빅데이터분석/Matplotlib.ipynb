### Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,6,7]
y=[20,30,50]
```

```python
plt.plot(x,y)
```

```python
x=np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x,np.sin(x));
```

```python
ax=plt.axes() #ax라는 객체 만듬
```

```python
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x),color='red',linestyle='solid')
plt.plot(x,np.cos(x),color='g',linestyle='--')
plt.title('sin&cos curves')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x,np.sin(x),color='red',linestyle='solid' ,label='sin(x)')
plt.plot(x,np.cos(x),color='g',linestyle='--', label='cos(x)')
plt.title('sin&cos curves')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.ylim((-1.2,1.2))
plt.legend()
plt.tight_layout() #이상한 수식 없애기
```

```python
np.random.seed(13)
x=np.random.random(20)
y=np.random.random(20)
```

```python
plt.plot(x,y,'s') #산점도 표시해야함
```

```python
plt.scatter(x,y) #산점도 표시 없어도됨
```

```python
np.random.seed(1)
dat=np.random.randn(1000)
```

```python
plt.hist(dat)
```

```python
np.random.seed(2)
x1=np.random.normal(0,0.8,1000)
x2=np.random.normal(-2,1,1000)
```

```python
plt.hist(x1,alpha=0.5,bins=30)#alpha 투명도 설정
plt.hist(x2,alpha=0.5,bins=30) #bins 히스토그램이 그려지는 각 칸의 개수 설정
```

```python
plt.subplot(2,3,1) #2*3=6개 6개를 그림
plt.plot(np.random.random(10))
plt.subplot(2,3,2) #2번
plt.plot(np.random.random(10))
plt.subplot(2,3,6) #6번
plt.plot(np.random.random(10))
```
